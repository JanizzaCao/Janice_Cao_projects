# 基础处理

```python
# 基础处理
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
path = "listings.csv"
df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

```python
df.neighbourhood.replace({'丰台区 / Fengtai':"丰台区",
                            "大兴区 / Daxing":"大兴区",
                            "密云县 / Miyun":"密云县",
                           "平谷区 / Pinggu":"平谷区",
                           "延庆县 / Yanqing":"延庆县",
                           "怀柔区 / Huairou":"怀柔区",
                           "朝阳区 / Chaoyang":"朝阳区",
                           "通州区 / Tongzhou":"通州区",
                           "门头沟区 / Mentougou":"门头沟区",
                           "顺义区 / Shunyi":"顺义区"},inplace = True)
```

## 截取需要列df_cut

```python
##截取需要列
df_cut = df[['id', 'host_id','neighbourhood',
             'room_type','price','minimum_nights','number_of_reviews',
             'last_review','reviews_per_month','calculated_host_listings_count']]
df_cut = df_cut.rename(columns = {'neighbourhood':'nb',
                                  'number_of_reviews':'n_of_r',
                                  'minimum_nights':'min_nights',
                                  'reviews_per_month':'r_per_m',
                                  'calculated_host_listings_count':'host_count',
                                  'last_review':'last_r'})
df_cut.head()
#df_cut.to_csv('beijing_airbnb_cut.csv')
```

```python
###各城区大于3QR的房产数量
price_innb_entire_75 = price_innb_entire_grouped.describe().iloc[:,6]
def more_75(nb):
    ####城区大于3QR的房产数量
    if nb == '东城区':
        x = 0
    elif nb == '丰台区':
        x = 1
    elif nb == '朝阳区':
        x = 2
    elif nb == '海淀区':
        x = 3
    else:
        x = 4
    more = price_innb[(price_innb.nb == nb)&
                  (price_innb.price > price_innb_entire_75[x])].reset_index(drop = True)
    amount = more.id.count()
    return amount
###各房价大于3QR的房产数量
nb = ['东城区','丰台区','朝阳区','海淀区','西城区']
for i in range(len(nb)):
    amount = more_75(nb[i])
    print("{}超出3QR房价{}元的数量为{}个".format(nb[i],price_innb_entire_75[i],amount))
```

## 房屋供应数量与区

```python

amount_grouped = df_amount.groupby(['nb','room_type']).id.count().unstack()
amount_grouped
```

```python
###求各区房产总数
amount_grouped['total'] = amount_grouped.apply(lambda x: x.sum(),axis = 1)
amount_grouped.sort_values(by = 'total', ascending = False, inplace = True)
amount_grouped
```

```python
###求各区房产类型占各区比例
amount_grouped['entire_pct'] = 100*round(amount_grouped['Entire home/apt']/amount_grouped['total'],4)
amount_grouped['private_pct'] = 100*round(amount_grouped['Private room']/amount_grouped['total'],4)
amount_grouped['shared_pct'] = 100*round(amount_grouped['Shared room']/amount_grouped['total'],4)
amount_grouped
amount_grouped.index
```
