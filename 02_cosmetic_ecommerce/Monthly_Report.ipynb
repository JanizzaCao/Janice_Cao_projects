# customer behavior analysis
## PV and UV
### daily for all events

```python
daily_PV = pd.read_excel(file1, sheet_name="2.11dailyPV")
daily_UV = pd.read_excel(file1, sheet_name="2.11dailyUV")
for x in files:
    dataPV = pd.read_excel(x, sheet_name="2.11dailyPV")
    daily_PV = pd.concat([daily_PV, dataPV])
    dataUV = pd.read_excel(x, sheet_name="2.11dailyUV")
    daily_UV = pd.concat([daily_UV, dataUV])
```

```python
daily_PV = daily_PV.reset_index(drop=True)
daily_UV = daily_UV.reset_index(drop=True)
```

```python
#max daily pv in each month
daily_PV["ym"] = daily_PV.date.dt.strftime("%Y-%m")
daily_PV["day"] = daily_PV.date.dt.strftime("%d")
#PV_max_index = daily_PV.groupby("ym").pv.rank(method="max")
PV_max_index = daily_PV.groupby("ym").pv.idxmax()
daily_PV_max = daily_PV.iloc[PV_max_index]
daily_PV_max
```

```python
daily_UV["ym"] = daily_UV.date.dt.strftime("%Y-%m")
daily_UV["day"] = daily_UV.date.dt.strftime("%d")
#PV_max_index = daily_UV.groupby("ym").uv.rank(method="max")
UV_max_index = daily_UV.groupby("ym").uv.idxmax()
daily_UV_max = daily_UV.iloc[UV_max_index]
daily_UV_max
```

```python
#plot the daily PV and daily UV
ym = daily_PV.ym.unique()
xlabel = range(1,32,2)

fig = plt.figure(figsize = (10,6))
sns.set(style = 'white', palette = 'muted')
ax = fig.add_subplot()
for month in ym:
    plot_data = daily_PV[daily_PV.ym == month]
    max_data = daily_PV_max[daily_PV_max.ym == month]
    plt.plot(plot_data.day, plot_data.pv, label=month)
    #plt.fill_between(plot_data.day,0,plot_data.pv )
    plt.plot(max_data.day, max_data.pv, marker = '*', markersize = 15)
    #maxdate = pv_d_max.iloc[0,2]
    #maxpv_d = pv_d_max.iloc[0,1]
    #plt.text(pv_d_max.day, pv_d_max.pv, 
    #         "   {} times in {}/10".format(maxpv_d, maxdate))
plt.title('Daily PV', fontsize = 18)
plt.xlabel('Day',fontsize = 15)
ax.set_yticks(range(50000,250001,50000))
ax.set_yticklabels(['50k','100k','150k','200k','250k'])
ax.set_xticks(['01','04','07','10','13','16','19','22','25','28','31'])
plt.grid(linestyle = "--", alpha = 0.5)
plt.legend()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
fig = plt.figure(figsize = (10,6))
sns.set(style = 'white', palette = 'bright')
ax = fig.add_subplot()
for month in ym:
    plot_data = daily_UV[daily_UV.ym == month]
    max_data = daily_UV_max[daily_UV_max.ym == month]
    plt.plot(plot_data.day, plot_data.uv, label=month)
    #plt.fill_between(plot_data.day,0,plot_data.pv )
    plt.plot(max_data.day, max_data.uv, marker = '*', markersize = 15)
    #maxdate = pv_d_max.iloc[0,2]
    #maxpv_d = pv_d_max.iloc[0,1]
    #plt.text(pv_d_max.day, pv_d_max.pv, 
    #         "   {} times in {}/10".format(maxpv_d, maxdate))
plt.title('Daily PV', fontsize = 18)
plt.xlabel('Day',fontsize = 15)
ax.set_yticks(range(5000,36000,5000))
ax.set_yticklabels(['5k','10k','15k','20k','25k','30k','35k'])
ax.set_xticks(['01','04','07','10','13','16','19','22','25','28','31'])
plt.grid(linestyle = "--", alpha = 0.5)
plt.legend()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

### Daily GMV

```python
month_average = {"month":[], "avg_GMV":[]}
for month, file in zip(ym, all_files):
    GMV_data = pd.read_excel(file, sheet_name="2.22dailyGMV")
    avg = sum(GMV_data.price)/len(GMV_data.price)
    month_average["month"].append(month)
    month_average["avg_GMV"].append(avg)
    print(GMV_data["price"].describe())

    fig = plt.figure(figsize=(10,6))
    ax = fig.add_subplot()
    sns.set(style = 'white', palette = 'bright')
    plt.plot(GMV_data["date"],GMV_data["price"])
    plt.grid(linestyle = '--', alpha = 0.5)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_xticklabels(range(1,GMV_data.date.nunique()+1,4))
    plt.title("Daily GMV in {}".format(month), fontsize=18)
    plt.xlabel("Date", fontsize=15)
    plt.ylabel("Total GMV", fontsize=15)
    plt.show()
```

```python
from datetime import timedelta
def retention_rate(user_data, n):
    """求所有用户n日留存率"""
    date_list = []
    retention_dayn = []
    date1 = pd.Series(user_data.index[:-n])
    for date in date1:
        daten = date+timedelta(n)
        user_dayone = user_data[user_data.index == date].user_id[0][1:-2].split(" ")
        """处理新用户
        new_user = set
        """
        user_dayn = user_data[user_data.index == daten].user_id[0][1:-2].split(" ")
        day_first = len(user_dayone)
        day_n = 0
        for user in user_dayone:
            if user in user_dayn:
                day_n += 1
        date_list.append(date.strftime("%d-%m-%y"))
        retention = round(day_n/day_first,4)
        retention_dayn.append(retention)
    return date_list, retention_dayn
```

```python
day_list2, retention_2days = retention_rate(user_num, 2)
day_list7, retention_7days = retention_rate(user_num, 7)
day_list30, retention_30days = retention_rate(user_num, 30)
```

## User 5 Levels Score For RFM model

```python
rfm_data5 = rfm.copy()
```

```python
rfm_data5['recent_value'] = pd.qcut(rfm_data5.recent_pur_time, 5, 
                                    labels = ['1',"2","3","4","5"])
rfm_data5['freq_value'] = pd.qcut(rfm_data5.pur_freq,5,
                                 labels = ['1',"2","3","4","5"])
rfm_data5['mon_value'] = pd.qcut(rfm_data5.total_momentary, 5, 
                                labels = ['1',"2","3","4","5"])
rfm_data5['rfm'] = rfm_data5['recent_value'].str.cat([rfm_data5.freq_value,rfm_data5.mon_value])
rfm_data5['recent_value'] = pd.to_numeric(rfm_data5['recent_value'])
rfm_data5['mon_value'] = pd.to_numeric(rfm_data5['mon_value'])
rfm_data5['freq_value'] = pd.to_numeric(rfm_data5['freq_value'])
rfm_data5['fm'] = (rfm_data5['mon_value']+rfm_data5['freq_value'] )/2
rfm_data5.head()
```

```python
def rfm_segment(r, fm):
    if fm <= 1:
        if r <=2:
            return "Lost"
        elif r <=3:
            return "About to Sleep"
        elif r <=4:
            return "Promising"
        else:
            return "New Customers"
    elif fm <=2:
        if r<=1:
            return "Lost"
        elif r<=2:
            return "Hibernating"
        elif r<=3:
            return "About to Sleep"
        else:
            return "Potential Loyalist"
    elif fm<=3:
        if r <= 2:
            return "At Risk"
        elif r <=3:
            return "Need Attention"
        else:
            return "Potential Loyalist"
    elif fm<=4:
        if r <=2:
            return "At Risk"
        else:
            return "Loyal Customers"
    else:
        if r <= 1:
            return "Can't Lose Them"
        elif r <= 2:
            return "At Risk"
        elif r <=4:
            return "Loyal Customers"
        else:
            return "Champions"
```

```python
x = []
for r, fm in zip (rfm_data5.recent_value, rfm_data5.fm):
    x.append(rfm_segment(r, fm))
rfm_data5['segment'] = x
rfm_data5.head()
```

```python
customer_segment = rfm_data5.groupby('segment').agg({"user_id":"count",
                                                    "total_momentary":"mean",
                                                    "pur_freq":"mean",
                                                    "gap":"mean"})
customer_segment
```

```python
#add segment percent:
customer_segment['percent'] = customer_segment.user_id.agg(lambda x: x/x.sum())
customer_segment
```

```python
fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot()
sns.set(style = "white", palette = "pastel")
plt.barh(range(len(customer_segment)),customer_segment.user_id)
plt.yticks(range(len(customer_segment)), labels = customer_segment.index)
plt.grid(linestyle = '--', alpha = 0.6)
plt.xlabel('Amount of Customers', fontsize =15)
plt.title('Amount of Customers in Each Segment', fontsize = 18)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
#plt.savefig("6.3 Amount of Customer in each segment.png")
```
